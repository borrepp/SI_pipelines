```python
from yaulab_processing import write_curatedFolder_into_nwb_sessions
```

```python
#########################################################################################################################
#                                              PREPROCESSING PATHS - SETTINGS
#########################################################################################################################

###############################################################################################################################################################
# PREPRO PARENT FOLDER INFORMATION:
###############################################################################################################################################################
parentFolder_preproNWB = "Y:\\Data_Albus\\Albus_NeuralData\\Albus-S1_preproNWB"

###############################################################################################################################################################
# THE CURATION FOLDER PATH from where to extract RECORDING & SORTING
# If set to None, it will prompt a window to ask user to select a PHY/Sorting_Analyzer folder
# Example curated from PHY:
# curatedFolder_path = "M:\\AlbusSorting_Results\\exp2024-04-03_sorting\\exp2024-04-03-sess-0_3-raw-B1-129-160\\exp2024-04-03-sess-0_3-raw-B1-129-160_MS5s2ch1thN4_5_phy"
# Example curated from Sorting_Analyzer:
# curatedFolder_path = "M:\\AlbusSorting_Results\\exp2024-04-03_sorting\\exp2024-04-03-sess-0_3-raw-D1-385\\exp2024-04-03-sess-0_3-raw-D1-385_MS5s2ch0th4_curated"
###############################################################################################################################################################
curatedFolder_path = None

###########################################################################################################################################
# GENERAL PARAMETER FOR "WAVEFORM TEMPLATES & LOCATIONS"
# The best practice is to match these values to the ones used during sorting.
ms_before = 0.6 # Default: 0.6
ms_after = 1.5 # Default: 1.5

###########################################################################################################################################
# start_unit_id = integer # it will reset the unitID with consecutive numbers, starting at variable value
# start_unid_id = None # it will use the unitID from PHY/sorting_analyzer as "unit_name"
start_unit_id = None # Default = None

###########################################################################################################################################
# Whether or not to ADD waveforms from each spike time. 
# This information will be extracted from the recording already imported to the NWB at:
# "nwb.processing.ecephys.ephys-ElectrodeGroup" 
add_waveforms = True

###########################################################################################################################################
# If the curated folder has been already splitted into sessions and exported as sorting_analyzer(s)
# it will read the existing temporal folders instead of computing the sorting_analyzer again. 
# Useful in case something goes wrong while exporting to NWB and there is no need to re-compute it.
rewriteAnalyzer = False

###########################################################################################################################################
# rewriteNWB_recording=False, If the recording has been already written into the NWB, it will be skipped. 
# rewriteNWB_recording=True, If the recording has been already written into the NWB, it will be removed and rewrite it. 
rewriteNWB_recording = False

###########################################################################################################################################
# rewriteNWB_units=False, If the units table has been already written into the NWB, it will be skipped. 
# rewriteNWB_units=True, If the units table has been already written into the NWB, it will be removed and rewrite it. 
rewriteNWB_units = False

###########################################################################################################################################
# Whether or not to delete the recording(s) & sorting_analyzer(s) created per session, after exporting the recordings & units from all sessions
removeAnalyzer = True

###########################################################################################################################################
# Whether or not to delete the binary recording associated with the curated folder (the concatenated recording used to sort & curate)
removeRecording = True

###########################################################################################################################################
# Whether or not to delete the PHY or SortingAnalyzer folder associated with the curated sorting
# The curated PHY/Sorting_analyzer folder will not be deleted for now
removeCuration = False

###########################################################################################################################################
# Whether or not to print messages related to the exporting process
verbose=True
```

```python
write_curatedFolder_into_nwb_sessions(
    curatedFolder_path, 
    parentFolder_preproNWB, 
    start_unit_id = start_unit_id, 
    ms_before = ms_before, 
    ms_after = ms_after, 
    add_waveforms = add_waveforms,
    rewriteAnalyzer = rewriteAnalyzer,
    rewriteNWB_recording = rewriteNWB_recording,
    rewriteNWB_units = rewriteNWB_units, 
    removeAnalyzer = removeAnalyzer,
    removeRecording = removeRecording,
    removeCuration = removeCuration,
    verbose = verbose
    )
```
